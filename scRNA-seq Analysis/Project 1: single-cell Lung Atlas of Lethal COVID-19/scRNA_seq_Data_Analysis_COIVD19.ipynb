## Doublet Removal

When you are making single cell libaries, somteimes two or more cells can end up in the same drop.

```python
!pip install scvi-tools
```

```python
import scvi
```

```python
adata #shows the dimensions of the cells by genes
```

will need to filter down the genes and test if a cell is a doublet or not, so lets narrow the number of genes (34546)

```python
sc.pp.filter_genes(adata, min_cells=10) #keep the genes that are found in at least 10 cells
adata
```

we have reduced the count of genes to 19896

```python
#keep the top 2000 variable genes
sc.pp.highly_variable_genes(adata, n_top_genes=2000, subset=True, flavor='seurat_v3')
```

```python
adata
```

observe that there are now only 2000 genes

```python
# model setup
scvi.model.SCVI.setup_anndata(adata)
# train vae model
vae = scvi.model.SCVI(adata)
vae.train()
```

```python
#train the solo model, which predicts doublets
solo = scvi.external.SOLO.from_scvi_model(vae)
```

```python
solo.train()
```

```python
solo.predict()
# cell barcode on the left
# score for doublet and singlet (higher scores represent prediciton)
```

```python
# pass soft=False to make a new column that is the predicted label (doublet or singlet)
df = solo.predict()
df['prediction'] = solo.predict(soft=False)

df
```

```python
# count how many doublets and singlets we have
df.groupby('prediction').count()
```

```python
# make a new column that is the difference in probability of doublet and singlet columns
df['diff'] = df.doublet - df.singlet
df
```

```python
import seaborn as sns
```

```python
sns.displot(df[df.prediction == 'doublet'], x='diff')
```

The histogram shown presents x-axis as the difference in prediction (doublet probability - singlet probability). The y-axis is the count of cell barcodes in that bin (difference). The histogram overall seems to be leveled, however, the distribution begins to drop dramatically than taper off at x = 0.4 This suggests that the cell barcodes at x < 0.4 can still likely be singlets, which we do not want to remove. So we will make the cutoff at x = 0.4 keeping 0.4 < x > 1.0; these are the doublets, with a new dataframe reload the raw cell barcode file and label the matching cell barcodes (from the processed dataframe) as doublets and eliminate them... thus keeping only singlets in the new dataframe.

```python
doublets = df[(df['prediction'] == 'doublet') & (df['diff'] > 0.4)]
doublets
```

```python
adata.obs['doublet'] = adata.obs.index.isin(doublets.index)
```

```python
adata.obs
```

```python
# ~ means you keep the 'False'
adata = adata[~adata.obs.doublet]
```

```python
adata
```

5610 cell barcodes that are singlets

doublets are now removed

## Preprocessing

```python
# mitochondrial genes are annotated as "MT-" in the gene names, lets filter that.
adata.var['mt'] = adata.var.index.str.startswith('MT-')
```

```python
adata.var
```

```python
# we need to label ribosomal genes
# we need to import a list of known ribosomal genes from the Broad Institute that can import into pandas
import pandas as pd
```

```python
ribo_url = "http://software.broadinstitute.org/gsea/msigdb/download_geneset.jsp?geneSetName=KEGG_RIBOSOME&fileType=txt"
ribo_genes = pd.read_table(ribo_url, skiprows=2, header=None)
```

```python
ribo_genes
```

```python
ribo_genes[0].values
```

```python
# same as we did with mitochondrial genes, we label it to our main dataframe
adata.var['ribo'] = adata.var_names.isin(ribo_genes[0].values)
```

```python
adata.var
```

```python
adata.obs
```

```python
# calculate qc metrics
sc.pp.calculate_qc_metrics(adata, qc_vars=['mt', 'ribo'], percent_top=None, log1p=False, inplace=True)
```

## Normalization

In single-cell RNA sequencing, there is alot of variation between cells, even between the same cell type due to sequencing biases, etc. Normaliztion will help compare cells and compare genes appropriately without the wide variations.

```python
adata.X.sum(axis=1) #total counts of cells, we need to normalize the counts for each cells so that their total counts adds up to the same value.
```

```python
sc.pp.normalize_total(adata, target_sum=1e4) # normalize every cell to 10,000 UMI
```

```python
adata.X.sum(axis=1) # each value (cell) got modified based on the starting number of counts for that cell.
```

```python
sc.pp.log1p(adata) #change to log counts
```

```python
adata.X.sum(axis=1)
```

## Clustering

```python
sc.pp.highly_variable_genes(adata, n_top_genes=2000)
```

```python
adata.var
```

```python
sc.pl.highly_variable_genes(adata)
```

```python
# Filter out the non-highly variable genes
adata = adata[:, adata.var.highly_variable]
```

```python
adata #we now have only the top 2000 genes that are highly variable
```

```python
# we're going to regress out the differences that arise due to the total number of counts mitochondrial counts and the ribosomal counts
#  this will filter out some data that are due to processing and just sample quality, sequencing artifact, etc.
sc.pp.regress_out(adata, ['total_counts', 'pct_counts_mt', 'pct_counts_ribo'])
```

```python
#normalize each gene to the unit variance of that gene
sc.pp.scale(adata, max_value=10)
```

```python
#Run component analysis to further reduce the dimensions of the data
sc.tl.pca(adata, svd_solver='arpack') # this calculates 50 pcs
```

```python
sc.pl.pca_variance_ratio(adata, log=True, n_pcs=50)
```

Plot of how much these pcs contribute. We want to look for where the data points tapers off to where values dont make a y-value difference anymore as you increase pcs value; this is at x = 25

```python
sc.pp.neighbors(adata, n_pcs=25)
```

```python
sc.tl.umap(adata)
```
